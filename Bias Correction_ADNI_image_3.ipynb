```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import shutil
import os
base_dir = '/mnt/data_lab513/vqtran_data'
#copy_data_dir = os.path.join(base_dir, "data", "Pre-Thesis_1")
count = 0
filename_list = []
for dirname, _, filenames in os.walk(root_data):
    for filename in filenames:
        #print(filename)
        #print(os.path.join(dirname, filename))
        path = os.path.join(dirname, filename)
        # load image
        ants_origin = ants.image_read(path)

        # bias correction
        ants_n3 = ants.n3_bias_field_correction(ants_origin)

        # save file
        bias_file = path.replace(root_data, root_bias_correction)

        # mkdir
        bias_dir = os.path.split(bias_file)[0]
        os.makedirs(bias_dir, exist_ok=True)

        count += 1
        # save file
        print("Save bias correction: {} - Progress: {}/{}".format(bias_file, count, len(os.listdir(root_data))))
        ants_n3.to_file(bias_file)
        
        
        
        
        
        #if ".nii" in filename:
        #    count += 1
        #    filename_list.append(filename)
            
        
#print(count)
#print(len(filename_list))
#filename_set = set(filename_list)
#print(len(filename_set))
print('complete bias correction')
```
